```python
import numpy as np
import pandas as pd

from pandas import DataFrame

data = {'names':['steve','john','richard','sarah','randy','michael','julie'],'age':[20,22,20,21,24,23,22],'gender':['male','male','male','female','male','male','female'],'rank':[2,1,4,5,3,7,6]}

ranking_df = DataFrame(data)
ranking_df.iloc[2:5,1]=np.nan
ranking_df.iloc[3:6,3]=np.nan
ranking_df.iloc[3,:]=np.nan
ranking_df

ranking_df.isnull()
ranking_df.notnull()
```

```python
ranking_df.dropna(how='all')
```

```python
ranking_df.isnull()
```

```python
bool_series=pd.isnull(ranking_df['age'])
ranking_df[bool_series]
```

```python
ranking_df.fillna(0)
```

```python
ranking_df.fillna(method='pad')
```

```python
ranking_df.fillna(method='bfill')
```

```python
ranking_df.interpolate(method='linear')
```
